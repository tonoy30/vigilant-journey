# Import 

```python
import os
import sys
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings;warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
```

```python
# set seed for reproducibility
np.random.seed(0)
```

# Data load

```python
data = pd.read_csv('../data/train.csv', index_col="id")
data.head()
```

# Spliting data into training and validation

```python
y = data.label
X = data.drop(['label'], axis=1)
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

# Removing missing values

```python
# drop cloumn from train and validation datasets if all the value is NaN
X_train.dropna(axis=1, how='all', inplace=True)
X_valid.dropna(axis=1, how='all', inplace=True)
```

```python
X_train.drop(["s54", "s55"], axis=1, inplace=True)
X_valid.drop(["s54", "s55"], axis=1, inplace=True)
```

```python
# find any ferther missing values
train_missing = X_train.isnull().sum()
valid_missing = X_valid.isnull().sum()
print("Missing from train dataset:", train_missing)
print("\n\nMissing from valid dataset:", valid_missing)
```

# Divide catagorical and non-catagorical data

```python
# categorical cols from train datasets
obj_train = (X_train.dtypes == 'object')
obj_train_cols = list(obj_train[obj_train].index)

print("Training - Categorical cols:\n", obj_train_cols)

obj_valid = (X_valid.dtypes == 'object')
obj_valid_cols = list(obj_valid[obj_valid].index)

print("\n\nValidation - Categorical cols:\n", obj_valid_cols)
```

```python
# non categorical cols from train datasets
non_obj_train = (X_train.dtypes != 'object')
non_obj_train_cols = list(non_obj_train[non_obj_train].index)

print("Training - non Categorical cols:\n", non_obj_train_cols)

non_obj_valid = (X_valid.dtypes != 'object')
non_obj_valid_cols = list(non_obj_valid[non_obj_valid].index)

print("\n\nValidation - non Categorical cols:\n", non_obj_valid_cols)
```

# Catagorical data

```python
# Apply ordinal encoder to each column with categorical data

from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
```

```python
X_train[obj_train_cols] = ordinal_encoder.fit_transform(X_train[obj_train_cols])
X_valid[obj_valid_cols] = ordinal_encoder.transform(X_valid[obj_valid_cols])
```

# Model

```python
from sklearn.ensemble import RandomForestClassifier

forest_model = RandomForestClassifier(random_state=0)
forest_model.fit(X_train, y_train)
```

```python
y_pred = forest_model.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_valid, y_pred)
```
